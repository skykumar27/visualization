```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig = plt.figure()
ax = plt.axes()

x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x));
```

```python
fig = plt.figure()
ax = plt.axes()

x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.cos(x));
```

```python
fig = plt.figure()
ax = plt.axes()

x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x));
ax.set_title('Simple plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig = plt.figure()
ax = plt.axes()

x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x));
ax.plot(x,np.cos(x));
ax.set_title('Multiple lines ')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x), label ='sin',color = 'black');
ax.plot(x,np.cos(x),label = 'cos', color = 'violet');
ax.set_title('Multiple lines ')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x), label ='sin',color = 'black', linestyle ='dashed');
ax.plot(x,np.cos(x),label = 'cos', color = 'violet', linestyle='dotted');
ax.plot(x,np.cos(x),label = 'cos', color = 'violet', linestyle='dashdot');
ax.set_title('Multiple lines ')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x), label ='sin');
ax.plot(x,np.cos(x),label = 'cos');
ax.plot(x,np.tan(x), label ='tan');
ax.set_title('Multiple lines ')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```
